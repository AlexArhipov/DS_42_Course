```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/ab-test.csv')
df
```

```python
fig, ax = plt.subplots(1, 2, sharey=True,figsize=(15, 8))
df[df.time == 'after'].boxplot(by='group', ax=ax[0],
                                  color=dict(boxes='green', whiskers='red',
                                             medians='red', caps='red'))
df[df.time == 'before'].boxplot(by='group', ax=ax[1],
                                   color=dict(boxes='green', whiskers='red',
                                              medians='red', caps='red'))
ax[0].set_title('after')
ax[1].set_title('before')
fig.suptitle(
    'Lag before labs deadlines. After and before the Newsfeed', fontsize=15)
plt.show()
```

What was the IQR of the control group before Newsfeed?

```python
iqr = df[(df.time == 'before') & (df.group == 'test')].quantile(.75) - df[(df.time == 'before') & (df.group == 'test')].quantile(.25)
round(iqr / 10) * 10
```
